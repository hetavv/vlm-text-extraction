```python
# Cell 11: Import Libraries and Define Paths
import os
import json
from tqdm.auto import tqdm
import jiwer # For WER and CER

# --- Define Base Paths ---
# Base path for your project within Google Drive
project_base_path = '/content/drive/MyDrive/VLM_Assignment'

# VLM Output Folders
raw_vlm_output_folder = os.path.join(project_base_path, 'vlm_raw_image_output')
enhanced_vlm_output_folder = os.path.join(project_base_path, 'vlm_gan_enhanced_image_output')

# Ground Truth Folder (flat structure as discussed)
ground_truth_folder = os.path.join(project_base_path, 'ground_truth_cleaned') # Using the cleaned GT

# --- Get List of All Image Basenames (e.g., 'Swarupsannidhan_0001') ---
# We can derive this from the raw VLM output folder, as it should contain all 100 images
# and their names directly correspond to the ground truth (minus the .txt extension)
image_basenames = []
for filename in os.listdir(raw_vlm_output_folder):
    if filename.endswith('.txt'):
        # Get the base name without the .txt extension
        base_name = os.path.splitext(filename)[0]
        image_basenames.append(base_name)

# Sort them to ensure consistent order
image_basenames.sort()

print(f"Found {len(image_basenames)} image basenames for evaluation.")

# --- Prepare Storage for Metrics ---
# Lists to store individual WER/CER/Custom scores for each image
wer_raw_list = []
cer_raw_list = []
wer_enhanced_list = []
cer_enhanced_list = []
custom_score_raw_list = []
custom_score_enhanced_list = []
```

```python
# Cell 12: Define Normalization Function for WER/CER
# This function prepares text by standardizing it before comparison.
# This is crucial for accurate WER/CER as it ignores irrelevant differences (e.g., extra spaces, punctuation).

def normalize_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove punctuation (you might want to customize this based on Gujarati script nuances)
    # This regex matches any character that is NOT a letter (Unicode) or a digit or a space
    import re
    text = re.sub(r'[^\w\s]', '', text, re.UNICODE)
    # Replace multiple spaces with a single space
    text = re.sub(r'\s+', ' ', text)
    # Strip leading/trailing whitespace
    text = text.strip()
    return text

# Example usage (optional, for testing)
# print(normalize_text("  હેલો, દુનિયા! (Hello, World!) ")) # Expected: "હેલો દુનિયા hello world"
```

```python
# Cell 13: Calculate WER and CER for Raw vs. GT and Enhanced vs. GT

print("\nStarting WER/CER calculation...")

for basename in tqdm(image_basenames, desc="Calculating WER/CER"):
    # Construct file paths for current image
    raw_vlm_filepath = os.path.join(raw_vlm_output_folder, f"{basename}.txt")
    enhanced_vlm_filepath = os.path.join(enhanced_vlm_output_folder, f"{basename}_restored.txt") # Assuming _restored suffix
    ground_truth_filepath = os.path.join(ground_truth_folder, f"{basename}.txt")

    # Read contents
    try:
        with open(ground_truth_filepath, 'r', encoding='utf-8') as f:
            reference = f.read()
        with open(raw_vlm_filepath, 'r', encoding='utf-8') as f:
            hypothesis_raw = f.read()
        with open(enhanced_vlm_filepath, 'r', encoding='utf-8') as f:
            hypothesis_enhanced = f.read()
    except FileNotFoundError as e:
        # Changed message to reflect likely cause of file not found
        print(f"Skipping WER/CER for {basename} due to non-existent ground truth file. Check path: {ground_truth_filepath}")
        continue
    except Exception as e:
        print(f"Error reading files for {basename} for WER/CER: {e}")
        continue

    # Normalize texts before calculating metrics
    reference_norm = normalize_text(reference)
    hypothesis_raw_norm = normalize_text(hypothesis_raw)
    hypothesis_enhanced_norm = normalize_text(hypothesis_enhanced)

    # --- Handle empty reference strings ---
    if not reference_norm:
        # Updated message for empty normalized reference
        print(f"Skipping WER/CER for {basename}: Normalized ground truth (reference) is empty, likely due to a blank/clear image or content removed by normalization.")
        continue # Skip to the next image

    # Calculate WER and CER
    # jiwer expects lists of strings, even for single sentences
    wer_raw = jiwer.wer(reference=[reference_norm], hypothesis=[hypothesis_raw_norm])
    cer_raw = jiwer.cer(reference=[reference_norm], hypothesis=[hypothesis_raw_norm])

    wer_enhanced = jiwer.wer(reference=[reference_norm], hypothesis=[hypothesis_enhanced_norm])
    cer_enhanced = jiwer.cer(reference=[reference_norm], hypothesis=[hypothesis_enhanced_norm])

    # Store individual scores
    wer_raw_list.append(wer_raw)
    cer_raw_list.append(cer_raw)
    wer_enhanced_list.append(wer_enhanced)
    cer_enhanced_list.append(cer_enhanced)

print("\nFinished WER/CER calculation.")
```

```python
# Cell 14: Custom Sequence-Accuracy Score Definition and Calculation

def calculate_custom_score(reference_text, hypothesis_text):
    """
    This example calculates a basic line-match accuracy.
    """
    ref_lines = [line.strip() for line in reference_text.split('\n') if line.strip()]
    hyp_lines = [line.strip() for line in hypothesis_text.split('\n') if line.strip()]

    if not ref_lines: # Handle empty ground truth
        return 1.0 if not hyp_lines else 0.0

    matched_lines = 0
    # Simple direct line comparison
    for ref_line in ref_lines:
        if ref_line in hyp_lines: # Check if the exact reference line exists in hypothesis
            matched_lines += 1

    # You could also try fuzzy matching, or order-aware comparison here.
    return matched_lines / len(ref_lines) # Percentage of reference lines found

print("\nStarting Custom Score calculation...")

for basename in tqdm(image_basenames, desc="Calculating Custom Scores"):
    # Construct file paths for current image
    raw_vlm_filepath = os.path.join(raw_vlm_output_folder, f"{basename}.txt")
    enhanced_vlm_filepath = os.path.join(enhanced_vlm_output_folder, f"{basename}_restored.txt") # Assuming _restored suffix
    ground_truth_filepath = os.path.join(ground_truth_folder, f"{basename}.txt")

    # Read contents (re-reading is fine, or you could pass them from Cell 13 if optimized)
    try:
        with open(ground_truth_filepath, 'r', encoding='utf-8') as f:
            reference = f.read()
        with open(raw_vlm_filepath, 'r', encoding='utf-8') as f:
            hypothesis_raw = f.read()
        with open(enhanced_vlm_filepath, 'r', encoding='utf-8') as f:
            hypothesis_enhanced = f.read()
    except FileNotFoundError as e:
        # Already handled in Cell 13, but good to be robust
        print(f"Skipping custom score for {basename} due to missing file: {e}")
        continue
    except Exception as e:
        print(f"Error reading files for {basename} for custom score: {e}")
        continue

    # Calculate custom score
    custom_raw = calculate_custom_score(reference, hypothesis_raw)
    custom_enhanced = calculate_custom_score(reference, hypothesis_enhanced)

    # Store individual scores
    custom_score_raw_list.append(custom_raw)
    custom_score_enhanced_list.append(custom_enhanced)

print("\nFinished Custom Score calculation.")
```
